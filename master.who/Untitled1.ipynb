```julia
using Vega
function main(window)
    push!(window.assets,("Vega","vega-plot"))
    
    x = 1:5
    y = rand(5)
    
    a = areaplot(x = x, y = y)
    colorscheme!(a, palette = "green") 
    b = barplot(x = x, y = y) 

    hbox(a, b) |> pad(2em)

end
```
